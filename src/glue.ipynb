Code to glue code from assemblies csv's

```python
import os
import pandas as pd
```

```python
# Some constants that might change depending on where spreadsheets are located
FINAL_FOLDER = '../csv/assemblies'
```

```python
# This code may need to be altered depending on how spreadsheets are saved
df = pd.DataFrame()
for file_name in os.listdir(FINAL_FOLDER):
  f = pd.read_csv(f'{FINAL_FOLDER}/{file_name}')
  f.insert(6, 'Term', file_name.split('_')[2][:-4])
  f.insert(7, 'Body', (['house'] if 'house' in file_name else ['senate']) * len(f))
  print(f)
  df = pd.concat([df, f], ignore_index=True)
df.to_csv(f'{FINAL_FOLDER}/../final.csv', index=False)
```
